```python
class Fan:
    def __init__(self):
        self.slow=1
        self.medium=2
        self.fast=3
        self.speed=self.slow
        self.radius=5
        self.color='blue'
        self.on='down'
        
    def getspeed(self):
        return self.speed
    def geton(self):
        return self.on
    def getradius(self):
        return self.radius
    def getcolor(self):
        return self.color
    def setspeed(self,speed):
         self.speed=speed
    def seton(self,on):
        self.on=on
    def setradius(self,radius):
        self.radius=radius
    def setcolor(self,color):
        self.color
    def change(self,speed,radius,color,on): 
        self.speed=speed
        self.radius=radius
        self.color=color
        self.on=on
    def show(self):
        print(self.speed,self.radius,self.color,self.on)

fan_1=Fan()
fan_1.change(3,10,'yellow','on')
fan_1.show()

fan_2=Fan()
fan_2.change(2,5,'blue','down')
fan_2.show()
```

- 8 进阶
- 定义一个子类，一个父类
    - 父类产生随机数
    - 子类一直判定该随机数是奇数还是偶数
        - 该例子实际上就是一个粗浅的分布式

```python
import random
class father:
    def __init__(self):
        self.Random = None
    def create_num(self):
        self.Random = random.randrange(0,100)
        return self.Random
class Son(father):
    def __init__(self):
        father.__init__(self)
    def result(self):
        while True:
            res_num = self.create_num()
            if res_num % 2 == 0:
                print("%d : 是偶数"%self.Random)
            else:
                 print("%d : 是奇数"%self.Random)
```

```python
Son().result()
```
